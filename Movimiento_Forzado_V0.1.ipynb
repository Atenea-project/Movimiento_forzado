# Práctica 3, Movimiento Forzado

## I) Teoría.

El movimiento armónico forzado (MAF) es el que describe a los cuerpos que oscilan o vibran y que se ven afectados por una fuerza externa que mantiene en oscilación al sistema. En estos casos el sistema siempre oscilara a la frecuencia de la fuerza aplicada aparte de una respuesta transitoria inicial, cabe destacar que la sensibilidad de la respuesta del sistema a la fuerza aplicada depende del factor de calidad.

En este cuadernillo veremos una descripción física del MAF con y sin amortiguamiento y repasaremos el concepto de resonancia.

    1) Movimiento armónico forzado sin amortiguar.
    2) Movimiento armónico forzado amortiguado.
    3) Resonancia.

### 1) Movimiento armónico forzado sin amortiguar.

Consideremos una masa $m$ acoplada a un resorte, a dicha masa se le aplica una fuerza externa $F=F_0\cos\omega t$ que la mantiene en oscilación, dado que la masa se encuentra bajo el efecto de una fuerza restitutiva del resorte y la fuerza aplicada la ecuación de Newton que describe dicho sistema es

\begin{equation}
    m\frac{d^2x}{dt^2}+kx=F_0\cos\omega t
\end{equation}

la solución a esta ecuación diferencial se describe con más detalle por (King, 2009), pero para fines prácticos nos podemos quedar con que la solución es

\begin{equation}
    x=A(\omega)\cos(\omega t-\delta)
\end{equation}

donde:

$A(\omega)$ es la amplitud física que podemos observar 

$\delta$ es un ángulo de fase entre la fuerza aplicada y el desplazamiento resultante. 

podemos deducir que $A(\omega)$ tiende a $a=F_0/k$ conforme $\omega$ tiende a cero, $A(\omega)$ tiende a cero conforme $\omega$ tiende a infinito y $A(\omega)$ tiende a infinito conforme $\omega$ tiende a $\omega_0$.

Para contemplar todo el espacio de frecuencias podemos reescribir la solución como

\begin{equation}
    x=C(\omega)\cos(\omega t)
\end{equation}

donde 

\begin{equation}
    C(\omega)=\frac{a}{\left(1-\frac{\omega^2}{\omega^2_0}\right)}
\end{equation}

### 2) Movimiento armónico forzado amortiguado.

En este caso la ecuación que describe este movimiento es 

\begin{equation}
    \frac{d^2x}{dt^2}+\gamma\frac{dx}{dt}+\omega_0^2x=\frac{F_0}{m}\cos(\omega t)
\end{equation}
Tal que:
\begin{align*}
    \gamma=\frac{b}{m}\\
    \omega_0^2=\frac{k}{m}
\end{align*}

Para solucionar esta ecuación diferencial tenemos que recurrir a los números complejos, con la relación de Euler.

\begin{align*}
    e^{i\omega t}=\cos(\omega t)+i\sin(\omega t)
\end{align*}

Con esto podemos pasar la ecuación diferencial al plano complejo.

\begin{equation}
    \frac{d^2Z}{dt^2}+\gamma\frac{dZ}{dt}+\omega_0^2Z=\frac{F_0}{m}e^{i\omega t}
\end{equation}

y proponemos la solución.

\begin{equation}
    Z=Ae^{i(\omega t - \delta)}
\end{equation}

Al aplicar dicha solución a la ecuación diferencial, obtenemos lo siguiente.

$$-A\omega^2e^{i(\omega t-\delta)}+\gamma(i\omega Ae^{i(\omega t-\delta)})+\omega_0^2(Ae^{i(\omega t-\delta)})=\frac{F_0}{m}e^{i \omega t}$$

$$-A\omega^2e^{-i\delta}+\gamma i\omega Ae^{-i\delta}+\omega_0^2Ae^{-i\delta}=\frac{F_0}{m}$$

$$Ae^{-i\delta}(-\omega^2+\gamma i\omega +\omega_0^2)=\frac{F_0}{m}$$

$$A(\omega_0^2-\omega^2)+A(\gamma i\omega2)=\frac{F_0}{m}e^{i\delta}$$

$$A(\omega_0^2-\omega^2)+A(\gamma i\omega2)=\frac{F_0}{m}(\cos(\delta)+i\sin(\delta))$$

Con lo que nos quedan las siguientes ecuaciones.

$$A(\omega_0^2-\omega^2)=\frac{F_0}{m}\cos(\delta)$$

$$A(\gamma i\omega2)=\frac{F_0}{m}i\sin(\delta)$$

Si elevamos ambas ecuaciones al cuadrado, simplificamos y sumamos ambas ecuaciones, obtenemos:

$$A^2((\omega_0^2-\omega^2)^2+(\gamma\omega)^2)=\frac{F_0^2}{m^2}(\cos^2(\delta)+\sin^2(\delta))$$

Con lo que podemos obtener lo siguiente:

$$A^2((\omega_0^2-\omega^2)^2+\gamma^2\omega^2)=\frac{F_0^2}{m^2}$$

$$A_{\omega}=\frac{a\omega_0^2}{\left[(\omega_0^2-\omega^2)^2+\gamma^2\omega^2\right]^{1/2}}=\frac{F_0}{m\left[(\omega_0^2-\omega^2)^2+\gamma^2\omega^2\right]^{1/2}}$$

Con lo cual podemos realizar algunas aproximaciones.

A) Si $\omega\approx0$
$$A_\omega\approx\frac{F_0}{m\omega_0^2}\approx\frac{F_0}{k}$$

B) Si $\omega\rightarrow\infty$
$$A_\omega\rightarrow0$$

C) Si $\omega\approx\omega_0$
$$A_\omega\approx\frac{F_0}{m\omega_0^2\gamma}\approx\frac{F_0}{k\gamma}$$

Con lo cual obtenemos las soluciones al estado estacionario para el movimiento forzado.

\begin{align}
    x=A_\omega(\cos(\omega t -\delta))\\
    v=-\omega A_\omega(\sin(\omega t -\delta))\\
    a=-\omega^2A_\omega(\cos(\omega t -\delta))
\end{align}

### 3) Resonancia.


La resonancia es un fenómeno en el que la amplitud de una oscilación alcanza un máximo. Esto ocurre cuando la frecuencia ($\omega$) de la fuerza aplicada tiende a la frecuencia natural ($\omega_0$) del sistema sobre el que se aplica la fuerza.

La resonancia ocurre cuándo un sistema es capaz de almacenar y transferir fácilmente energía entre dos o más formas de almacenamiento diferentes, pensemos en lo que pasa cuando una cantante de ópera hacer resonar una copa de vidrio. Sin embargo, hay algunas perdidas de ciclo a ciclo, llamado amortiguamiento. Cuando el amortiguamiento es pequeño, la frecuencia de resonancia es aproximadamente igual a la frecuencia natural del sistema, el cual es una frecuencia de vibraciones no forzadas. Algunos sistemas tienen múltiples frecuencias resonantes. 

En la gráfica que construiste puedes comprobar como la amplitud siempre tiende a un máximo conforme 'w' y 'w0' se aproximan.

#### 3.1) Resonancia en circuitos eléctricos.

En el caso de un circuito eléctrico del tipo RLC que tiene una fuente de poder conectada podemos aplicar el modelo de Movimiento armónico forzado, partimos de la ecuación 

$$L\frac{d^2q}{dt^2}+R\frac{dq}{dt}+\frac{q}{C}=V_0\cos(\omega t)$$

\begin{equation}\label{res_RLC}
    \frac{d^2q}{dt^2}+\gamma\frac{dq}{dt}+\omega_0^2q=\frac{V_0}{L}\cos(\omega t)
\end{equation}

recordando que

\begin{align*}
    \omega_0^2=\frac{1}{LC}\\
    \gamma=\frac{R}{L}\\
    Q=\frac{\omega_0}{\gamma}=\frac{1}{R}\sqrt{\frac{L}{C}}
\end{align*}

La función solución para la ecuación diferencial queda

\begin{equation}
    q=q_0(\omega)\cos(\omega t-\delta)
\end{equation}

Donde definimos

\begin{align*}
    q_0(\omega)=\frac{V_0/L}{[(\omega_0^2-\omega^2)^2+(R\omega/L)^2]^{1/2}}\\
    =\frac{V_0}{\omega[(1/\omega C-\omega L)^2+R^2]^{1/2}}
\end{align*}

Para este sistema podemos definir la corriente y la diferencia de potencial de la siguiente manera.

$$I=\frac{dq}{dt}=-q_0(\omega)\omega\sin(\omega t-\delta)$$

\begin{equation}
    I=\frac{-V_0\sin(\omega t-\delta)}{[(1/\omega C-\omega L)^2+R^2]^{1/2}}
\end{equation}

$$V_C=V_C(\omega)\omega\cos(\omega t-\delta)$$

\begin{equation}
    V_C=\frac{V_0/LC}{[(\omega_0^2-\omega^2)^2+(R\omega/L)^2]^{1/2}}
\end{equation}

En resonancia $\omega=\omega_0$ por lo que la ecuación anterior queda como:

\begin{equation}\label{DV_Resonancia_RLC0}
    V_C(\omega_0)=\frac{V_0}{R\omega_0C}=QV_0
\end{equation}
